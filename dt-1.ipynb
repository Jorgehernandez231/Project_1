```python
df = dt[dt['date'] >= "1930-01-01"]
df
```

```python
def get_winner(row):
    if row["home_score"] > row["away_score"]:
        return row["home_team"]
    elif row["home_score"] < row["away_score"]:
        return row["away_team"]
    else:
        return "Draw"  


df["winner"] = df.apply(get_winner, axis=1)


win_counts = df[df["winner"] != "Draw"]["winner"].value_counts()


top10_totalwinners = win_counts.head(10)

top10_totalwinners
```

```python
df_off = df[df["tournament"] != "Friendly"].copy()

df_off["tournament"].value_counts()
```

```python
df_off["off_winner"] = df_off.apply(get_winner, axis=1)


win_counts = df_off[df_off["off_winner"] != "Draw"]["off_winner"].value_counts()


top10_offwinners = win_counts.head(10)

top10_offwinners
```

```python
df_worldcup = df[df["tournament"] == "FIFA World Cup"].copy()

df_worldcup["tournament"].value_counts()
```

```python
df_worldcup["worldcup_winner"] = df_worldcup.apply(get_winner, axis=1)


win_counts = df_worldcup[df_worldcup["worldcup_winner"] != "Draw"]["worldcup_winner"].value_counts()


top10_worldcup_winners = win_counts.head(10)

top10_worldcup_winners
```

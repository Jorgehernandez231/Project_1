```python
def get_winner(row):
    if row["home_score"] > row["away_score"]:
        return row["home_team"]
    elif row["home_score"] < row["away_score"]:
        return row["away_team"]
    else:
        return "Draw"  


dt["winner"] = dt.apply(get_winner, axis=1)


win_counts = dt[dt["winner"] != "Draw"]["winner"].value_counts()


top10_totalwinners = win_counts.head(10)

top10_totalwinners
```

```python
dt_merge = pd.merge(dt, dt_p, on= ['date', 'home_team', 'away_team'], how = 'left')
dt_merge
```

```python
dt_merge['winner_x'] = dt_merge.apply(lambda row: row['winner_y'] if row['winner_x'] == 'Draw' else row['winner_x'], axis=1)
dt_merge
   
```

```python
dt_merge.winner_x.isna().sum()
```

```python
dt_merge['winner_x'] = dt_merge.winner_x.fillna('Draw')
dt_merge
```

```python
dt_merge = dt_merge.drop(columns=['winner_y','first_shooter']).rename(columns={'winner_x': 'winner'})
dt_merge
```
